```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("EPL Home Away xG xGA 23_24.xlsx")
df
```

```python
# team performance at home
home_perf = df[['Squad', 'home_MP', 'home_W', 'home_D', 'home_L', 'home_GF', 'home_GA', 'home_GD', 'home_Pts', 'home_Pts/MP']].copy()
home_perf['type'] = 'Home'

home_perf

```

```python
# performance of teams away from home
away_perf = df[['Squad', 'away_MP', 'away_W', 'away_D', 'away_L', 'away_GF', 'away_GA', 'away_GD', 'away_Pts', 'away_Pts/MP']].copy()
away_perf['type'] = 'Away'
away_perf
```

```python
away_perf.columns = home_perf.columns

combined_perf = pd.concat([home_perf, away_perf], ignore_index=True)
```

```python
# graph with points won at home and away
plt.figure(figsize=(14, 7))
sns.barplot(data=combined_perf, x='Squad', y='home_Pts', hue='type')
plt.title('points home vs away')
plt.xlabel('Tean')
plt.ylabel('Points')
plt.xticks(rotation=90)
plt.savefig('points-homexaway.png')
plt.show
```

```python
# graph with goals scored at home and away
plt.figure(figsize=(14, 7))
sns.barplot(data=combined_perf, x='Squad', y='home_GF', hue='type')
plt.title('Goals scored home and away')
plt.xlabel('Tean')
plt.ylabel('Goals scored')
plt.xticks(rotation=90)
plt.savefig('goals-scr-home-and-away.png')
plt.show()
```

```python
# goals expected vs goals scored. GX and GF
relaction_GF_xG = df['home_xG'].corr(df['home_GF'])

relaction_GA_xG = df['away_xG'].corr(df['away_GA'])

print(f"GX and GF: {relaction_GF_xG}")
print(f"xG and GA: {relaction_GA_xG}")
```

```python
# Create the scatter plot for xG vs GF 
plt.figure(figsize=(14, 7))
plt.scatter(df['home_xG'], df['home_GF'], color='blue', label='Goals Scored at Home')
plt.scatter(df['away_xG'], df['away_GF'], color='red', label='Goals Scored Away')
for i, row in df.iterrows():
    plt.annotate(row['Squad'], (row['home_xG'], row['home_GF']), textcoords="offset points", xytext=(5,-5), ha='left', va='bottom', fontsize=8, color='black', bbox=dict(facecolor='white', alpha=0.5))
    plt.annotate(row['Squad'], (row['away_xG'], row['away_GF']), textcoords="offset points", xytext=(5,-5), ha='left', va='bottom', fontsize=8, color='black', bbox=dict(facecolor='white', alpha=0.5))
plt.title('xG vs Goals Scored')
plt.xlabel('xG')
plt.ylabel('Goals Scored')
plt.legend()
plt.savefig('goals-scr-away.png')
plt.show()
```

```python
# Create the scatter plot for xG vs GA
plt.figure(figsize=(14, 7))
plt.scatter(df['home_xG'], df['home_GA'], color='blue', label='Goals Conceded at Home')
plt.scatter(df['away_xG'], df['away_GA'], color='red', label='Goals Conceded Away')
for i, row in df.iterrows():
    plt.annotate(row['Squad'], (row['home_xG'], row['home_GA']), textcoords="offset points", xytext=(5,-5), ha='left', va='bottom', fontsize=8, color='black', bbox=dict(facecolor='white', alpha=0.5))
    plt.annotate(row['Squad'], (row['away_xG'], row['away_GA']), textcoords="offset points", xytext=(5,-5), ha='left', va='bottom', fontsize=8, color='black', bbox=dict(facecolor='white', alpha=0.5))
plt.title('xG vs Goals Conceded')
plt.xlabel('xG')
plt.ylabel('Goals Conceded')
plt.legend()
plt.savefig('xGvsg.png')
plt.show()
```

```python
# best strike at home and away(more goals scored)
best_attack_h = df[['Squad', 'home_GF']].sort_values(by='home_GF', ascending=False).head(10)
best_attack_a = df[['Squad', 'away_GF']].sort_values(by='away_GF', ascending=False).head(10)
# best defense at home and away(less goals conseded)
best_defense_h = df[['Squad', 'home_GA']].sort_values(by='home_GA', ascending=True).head(10)
best_defense_a = df[['Squad', 'away_GA']].sort_values(by='away_GA', ascending=True).head(10)
```

```python
# graphics to make viewing easier
# first, bestter strike at home
plt.figure(figsize=(14, 7))
plt.bar(best_attack_h['Squad'], best_attack_h['home_GF'], color='green')
plt.title('Top 10 Best attack at Home in EPL')
plt.xlabel('Tean')
plt.ylabel('Goals scored at home')
plt.xticks(rotation=45)
plt.savefig('top10-goals-scr-home.png')
plt.show()
```

```python
# now, better attack away
plt.figure(figsize=(14, 7))
plt.bar(best_attack_a['Squad'], best_attack_a['away_GF'], color='green')
plt.title('Top 10 Best attack away in EPL')
plt.xlabel('Tean')
plt.ylabel('Goals scored away')
plt.xticks(rotation=45)
plt.savefig('top10-goals-scr-away.png')
plt.show()
```

```python
# defense graphics
# best home defense
plt.figure(figsize=(14,7))
plt.bar(best_defense_h['Squad'], best_defense_h['home_GA'], color='blue')
plt.title('Top best defense at home')
plt.xlabel('Tean')
plt.ylabel('Goals conceded at home')
plt.xticks(rotation=45)
plt.savefig('goals-cons-home.png')
plt.show()
```

```python
# best defense away
plt.figure(figsize=(14,7))
plt.bar(best_defense_a['Squad'], best_defense_a['away_GA'], color='blue')
plt.title('Top best defense away')
plt.xlabel('Tean')
plt.ylabel('Goals conceded away')
plt.xticks(rotation=45)
plt.savefig('goals-cons-away.png')
plt.show()
```

```python
# Add up wins, draws and defeats at home and away
home_stats = df[['Squad', 'home_W', 'home_D', 'home_L']].set_index('Squad').sum()
away_stats = df[['Squad', 'away_W', 'away_D', 'away_L']].set_index('Squad').sum()

categories = ['home_W', 'home_D', 'home_L', 'away_W', 'away_D', 'away_L']
values = [home_stats['home_W'], home_stats['home_D'], home_stats['home_L'], away_stats['away_W'], away_stats['away_D'], away_stats['away_L']]

#Plot the frequencies of wins, draws and losses at home
plt.figure(figsize=(14, 7))
plt.bar(df['Squad'], df['home_W'], color='green', label='Wins')
plt.bar(df['Squad'], df['home_D'], bottom=df['home_W'], color='yellow', label='Draws')
plt.bar(df['Squad'], df['home_L'], bottom=df['home_W'] + df['home_D'], color='red', label='Losses')
plt.title('Home: Wins, Draws, and Losses per Team')
plt.xlabel('Team')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend()
plt.savefig('freq-home.png')
plt.show()
```

```python
# Plot the frequencies of wins, draws and away defeats
plt.figure(figsize=(14, 7))
plt.bar(df['Squad'], df['away_W'], color='green', label='Wins')
plt.bar(df['Squad'], df['away_D'], bottom=df['away_W'], color='yellow', label='Draws')
plt.bar(df['Squad'], df['away_L'], bottom=df['away_W'] + df['away_D'], color='red', label='Losses')
plt.title('Away: Wins, Draws, and Losses per Team')
plt.xlabel('Team')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend()
plt.savefig('freq-away.png')
plt.show()
plt.close()
```
